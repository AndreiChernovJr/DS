## Лаги и средние скользящие

```python
#check = check.sort_values(by=['date', 'st_id', 'pr_sku_id'])
```

```python
for n in range(7, 71, 7):
    check[f'lag{n}'] = check.total_sales\
        .shift(check.date.value_counts().values[0]*n)
```

```python
for m in range(7, 36, 7):
    check[f'rolling_mean_{m}'] = check.total_sales.rolling(check.date.value_counts().values[0]*m).mean()
```

```python
train_data = check_enc[check_enc.date<='2023-07-04']
test_data = check_enc[check_enc.date>'2023-07-04']
```

```python
x_train = train_data.drop('total_sales', axis=1)
y_train = train_data['total_sales']
```

```python
#print(x_train.dtypes)
#print(y_train.dtype)
```

```python
x_test = test_data.drop('total_sales', axis=1)
y_test = test_data['total_sales']
```

```python
def wape(y_true, y_pred):
    return np.sum(np.abs(y_true-y_pred))/np.sum(np.abs(y_true))
```

```python
score = make_scorer(wape, greater_is_better=False)
```

```python
#x_train = x_train.drop('date', axis = 1)
x_train['date'] = pd.to_numeric(x_train['date'], errors='coerce')
```

```python
tscv = TimeSeriesSplit(n_splits=6)
lr = LinearRegression()
parametrs = {}
lr_grid = GridSearchCV(lr, parametrs, scoring=score, cv=tscv)
lr_grid.fit(x_train, y_train)
```

```python
x_test['date'] = pd.to_numeric(x_test['date'], errors='coerce')
```

```python
lr_preds = lr_grid.best_estimator_.predict(x_test)
```

```python
wape(y_test, lr_preds)
```
